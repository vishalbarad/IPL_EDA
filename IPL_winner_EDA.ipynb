## winning probabilities of each teams including all the season

```python
# win probabilities of each teams including all the season
sum = Match_origin['winner'].value_counts().sum()
team = list()
win_proba = list()
for i in range(len(Match_origin['winner'].value_counts())):
    team.append(Match_origin['winner'].value_counts().index[i])
    win_proba.append(Match_origin['winner'].value_counts()[i]/sum)

plt.figure(figsize=(10,6))
plt.barh(team,win_proba,color='r')
plt.title("Team win probabilities")
plt.xlabel('Probabilities')
plt.ylabel('Team')
```

## Season (2008-2017) wise winning probabilities

```python
# Season (2008-2017) wise win probabilities
def Match_win_proba(season):
    Match_origin[Match_origin['season']==season]['winner'].value_counts()
    s_sum = Match_origin[Match_origin['season']==season]['winner'].value_counts().sum()
    s_team = list()
    s_win_proba = list()
    
    for i in range(len(Match_origin[Match_origin['season']==season]['winner'].value_counts())):
        s_team.append(Match_origin[Match_origin['season']==season]['winner'].value_counts().index[i])
        s_win_proba.append(Match_origin[Match_origin['season']==season]['winner'].value_counts()[i]/sum)
    
    plt.figure(figsize=(10,6))
    plt.barh(s_team,s_win_proba,color='r')
    plt.title("Teams win probabilities of season {}".format(season))
    plt.xlabel('Probabilities')
    plt.ylabel('Team')
```

## Top 5 players's total runs of the team 

```python
def Bats_man_run(team):
    run = list()
    name = list()
    for bats_man in Delivery[Delivery['batting_team']==team]['batsman'].unique()[:5]:
        team_batting_score = Delivery[Delivery['batting_team']==team][['batsman','batsman_runs']]
        run.append(team_batting_score[team_batting_score['batsman']==bats_man]['batsman_runs'].sum())
        name.append(bats_man)
    d={}
    for run,name in zip(run,name):
        d[name] = run
    return d
```

### You can check for each team e.g following for the 'Chennai Super Kings' team

```python
team = 'Royal Challengers Bangalore'
run_dict = Bats_man_run(team)
for x, y in run_dict.items():
    print("Total IPL runs of '"+x+"' are ",y)
    plt.barh(x,y)
    plt.title("{}'s top 5 players total runs so far".format(team))
    plt.xlabel('Total Runs')
    plt.ylabel('Player name')
```

## Top 5 players's total wickets of the team

```python
def Bowler_wickets(team):
    wickets = list()
    name = list()
    for bowler in Bowler_profile[Bowler_profile['bowling_team']==team]['bowler'].unique()[:5]:
        team_bowler_wicket = Bowler_profile[Bowler_profile['bowling_team']==team][['bowler','player_dismissed']]
        wickets.append(team_bowler_wicket[team_bowler_wicket['bowler']==bowler]['player_dismissed'].sum())
        name.append(bowler)
    d={}
    for wickets,name in zip(wickets,name):
        d[name] = wickets
    return d
```

## Match result either Normal, Tie, No result between team1 and team2

```python
def Match_result(team1,team2):
    Normal = Match_origin[(Match_origin['team1']==team1) & (Match_origin['team2']==team2)]['result'].value_counts()
    plt.pie(list(Normal),labels=list(Normal.keys()),autopct='%0.1f%%')
    plt.legend(('Normal','Tie','Noresult'))
    plt.title("Match result between '{}' and '{}' ".format(team1,team2))
```

## How many matches team1 and team2 won in past number of matches

```python
def last_matches_won(team1,team2,last_matches):
    last_Match_played = Match_origin[(Match_origin['team1']==team1) & (Match_origin['team2']==team2)]
    last_numbers = last_Match_played.sort_values(by=['date'],ascending=False).head(last_matches)
    print("In past {} matches".format(last_matches))
    print("'{}' won {} mathces.".format(team1,last_numbers[last_numbers['winner']==team1].shape[0]))
    print("'{}' won {} mathces.".format(team2,last_numbers[last_numbers['winner']==team2].shape[0]))
    sns.countplot(last_numbers['winner'],data=last_numbers)
    plt.ylabel("Number os matches")
```

### Team performance on stadium ground

```python
def team_performance_stadium(team,stadium):
    Stadium_match_won = New_Matches[(New_Matches['venue']==stadium)]['winner']==team
    l = Stadium_match_won.value_counts()
    j=-1
    re=list()
    for i in l.keys():
        j+=1
        if i==True:
            re = l[j]
            break
        else:
            continue
    return "{} won {} matches on {}".format(team,re,stadium)
```
